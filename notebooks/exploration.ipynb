```python
import pandas as pd # Importamos la liberia pandas para manejar y analizar datos
import matplotlib.pyplot as plt # Importamos la libreria matplotlib para crear graficos

# Cargamos los datos limpios

data_path = "../data/procesado/cleaned_products.csv"
df = pd.read_csv(data_path) # Carga de datos de un archivo CSV en un DataFrame

# Analisis y graficos

print ("Basic Statistics")

print (df.describe()) #Mostramos un resumen estadistico de los datos en el DataFrame

#Histograma de precios

plt.hist(df["price"], bins=20, edgecolor="black") #Creamos un histograma de la columna "Price" con 20 barras y bordes negros 

plt.title("Distribucion de precios") #Añadimos un titulo al histograma
plt.xlabel("Precios") #Añadimos un label al eje x
plt.ylabel("Frecuencia") #Añadimos un label al eje y
plt.show() #Mostramos el histograma

#Productos mas caros

print("\nProductos mas caros:") #Imprimimos un encabezado para los productos con los precios mas altos
highestPrices = df.nlargest(5, "price") #Guardamos los 5 productos con los precios mas altos en una variable
print(highestPrices) #Imprimimos los 5 productos con los precios mas altos

```
